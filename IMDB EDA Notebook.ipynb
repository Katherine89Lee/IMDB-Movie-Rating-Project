# Finding a way to visualize the Overview column.

```python
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
moviedf.overview.head()
```

```python
text = " ".join(review for review in moviedf.overview.astype(str))
print ("There are {} words in the combination of all cells in column overview.".format(len(text)))
```

```python
stopwords = set(STOPWORDS)
```

```python
wordcloud = WordCloud(stopwords=stopwords, background_color="white", width=800, height=400).generate(text)
plt.axis("off")

plt.figure(figsize = (40, 20))

plt.tight_layout(pad=0)

plt.imshow(wordcloud, interpolation='bilinear')

plt.show()
```
